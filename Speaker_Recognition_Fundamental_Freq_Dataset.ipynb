```python
def freq_comp(dr):
    drlist=os.listdir(dr)
    music=[]
    ext=[".wav",".mp3"]
    for i in drlist:
        if i[-4:] in ext:
            music.append(i)
    all_song=[]
    lendata=[]
    for j in range(len(music)):
        x, sr = librosa.load(dr+music[j])
        X = scipy.fft(x[:4096])
        X_mag = numpy.absolute(X)
        f = numpy.linspace(0, sr, 4096)
        X_mag1=X_mag.copy()
        X_mag1.sort()
        temp=np.unique(X_mag1[::-1][:100])
        temp=temp[::-1]
        #print("%0.1f"%f[X_mag.tolist().index(temp[0])])
        #print(X_mag[temp[0]])
        freq_data=[]      # Contains range of voice fundamental frequncies for each Train Song
        
        for itr in range(len(temp)):
            if float("%0.1f"%f[X_mag.tolist().index(temp[itr])])>80:
                freq_data.append(float("%0.1f"%f[X_mag.tolist().index(temp[itr])]))
        lendata.append(len(freq_data))  
        all_song.append(freq_data)
    return all_song,lendata,music
```

```python
dr="D:/As Freelancer/Projects/Speaker Recognition/wave/TRAIN/"
all_song,lendata,music=freq_comp(dr)
name=[]
for i in music:
    name.append(i[:-6].strip('-'))
mean_len=int(np.array(lendata).mean())
print("Actual Length Data: ",lendata)
print("Mean Length: ",mean_len)
currlen=[]
for i in range(len(all_song)):
    if len(all_song[i])>mean_len:
        all_song[i]=all_song[i][:mean_len]
    elif len(all_song[i])<mean_len:
        diff=mean_len-len(all_song[i])
        for ins in range(diff):
            all_song[i].append(min(all_song[i]))
    currlen.append(len(all_song[i]))
print("Modified Length Data: ",currlen)
```
